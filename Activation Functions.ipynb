# Activation Function

### Activation functions help map input values to a known range, which helps stabilize training and helps map values to a desired output in the last layer. The most popular activation functions include  sigmoid, tanh, ReLU, and Leaky ReLU.

• Sigmoid: Squeezes values between 0 and 1, used for binary outputs.
    
• Tanh: Squeezes values between -1 and 1, better centered around zero.
    
• ReLU: Outputs positive values as-is, zeros out negatives, popular in deep learning.
    
• Leaky ReLU: Like ReLU but allows a small negative slope, preventing dead neurons.

```python
import math
```

```python
def sigmoid(x): # this coverts any value to range between 0 & 1
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(0.1)
```

```python
def tanh(x): # this coverts any value to range between 1 & -1
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(100)
```

```python
tanh(-56)
```

```python
tanh(1)
```

```python
def leaky_relu(x): # the negative section allows a small gradient instead of being completely zero
    return max(0.1*x,x)
```

```python
leaky_relu(100)
```

```python
leaky_relu(-10)
```
